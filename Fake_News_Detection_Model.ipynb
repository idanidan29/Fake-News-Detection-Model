```python
# 1. Install & Setup Backend
!pip install textblob
import os
os.environ["KERAS_BACKEND"] = "tensorflow"  # You can also use "jax" or "torch"

import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

# Scikit-Learn for Pre-processing & Evaluation
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve

# Keras 3 Imports
import keras
from keras import layers
from keras import ops  # Useful for multi-backend math
```

```python
# 1. Load the dataset from your sample_data folder
df = pd.read_csv('/content/sample_data/WELFake_Dataset.csv')

# 2. Basic Cleaning: Drop rows where text or title is missing
df.dropna(subset=['title', 'text'], inplace=True)

# 3. Create 'full_content' (The raw material for our analysis)
df['full_content'] = df['title'] + " " + df['text']

# 4. Noise Reduction Function (Requirement 2.1 in your README)
def clean_text(text):
    # Remove HTML tags (if any)
    text = re.sub(r'<.*?>', '', text)
    # Remove special characters and numbers (keeping only letters)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convert to lowercase
    text = text.lower().strip()
    return text

# 5. Apply cleaning to a sample to keep the notebook fast during development
# Note: WELFake is large (72k rows). We'll use 20k for now to ensure the GPU doesn't crash.
df_sample = df.sample(20000, random_state=42).copy()
df_sample['clean_text'] = df_sample['full_content'].apply(clean_text)

print(f"Dataset loaded. Sample size: {len(df_sample)}")
df_sample[['full_content', 'clean_text']].head(2)
```
